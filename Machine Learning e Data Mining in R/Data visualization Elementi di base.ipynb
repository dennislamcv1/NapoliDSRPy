## Data visualization: elementi di base

Attraverso i seguenti esempi, provo ad illustrarti, ad un livello crescente di complessità, le tipiche istruzioni utilizzate per diagrammare il contenuto delle variabili con i pacchetti di base di R

### Grafico di una variabile (numerica)

Carica il dataset `Auto`, contenuto anche nel pacchetto `ISLR`, e diagramma la colonna `mpg`. Sulle ordinate c'è il valore della variabile, sulle ascisse il numero di riga.

```r
df <- read.csv("data/Auto.csv")
plot(df$mpg)
```

```r
plot(df$mpg, type = "l") # Linee invece di punti 
```

```r
plot(df$mpg, col = "red") # Colore punti/linee
```

```r
plot(df$mpg, pch = 16) # Cambiare la forma dei punti
```

```r
plot(df$mpg, cex = 3) # Cambiare le dimensioni dei punti/linee
```

```r
plot(df$mpg, type = "l", lty = 2) # Cambiare il tipo delle linee
```

```r
plot(df$mpg, main = "Title") # Titolo del grafico
```

```r
plot(df$mpg, xlab = "x axis") # Etichetta dell'asse x
```

```r
plot(df$mpg, ylab = "y axis") # Etichetta dell'asse y
```

```r
plot(df$mpg, xlim = c(10, 50)) # Limiti dell'asse x
```

```r
plot(df$mpg, ylim = c(0, 60)) # Limiti dell'asse y
```

### Istogramma

L'istogramma è costituito da rettangoli adiacenti, le cui basi sono allineate su un asse orientato e dotato di unità di misura, che rappresenta l'asse delle ascisse. 
Ogni rettangolo ha base arbitraria pari all'ampiezza della corrispondente classe; l'altezza invece è calcolata come frequenza assoluta, pari al numero di osservazioni incluse nella corrispondente classe. 

densità di frequenza (data dal rapporto tra frequenza assoluta ed il prodotto tra l’ampiezza della classe ed il numero di osservazioni totali). Il comando per ottenere un istogramma è `hist()`

```r
hist(df$mpg) # istogramma delle frequenze
```

L'argomento `breaks=` permette di regolare il numero di classi. Prova a modificarlo.

```r
hist(df$mpg, breaks = 40) # numero di intervalli
```

L'argomento `probability = TRUE` è utile nel caso in cui volessi modificare la scala delle *y* dell'istogramma in modo che l'area totale di tutti i rettangoli sia pari ad 1. 

```r
hist(df$mpg, probability = TRUE) # istogramma delle densità
```

Per esempio, potresti voler rappresentare mediante l'istogramma `n`=1000 osservazioni di una v.a. normale standard

```r
hist(rnorm(1000), probability = TRUE) # istogramma delle densità
```

Tale argomento è utile nel caso in cui volessi sovrapporre all'istogramma il grafico di una pdf normale standard.
Per farlo, come hai visto nel reading precedente, usa il comando `curve` con l'argomento `add = TRUE`.

Gli argomenti aggiuntivi `xlim=` e `ylim=` modificano, rispettivamente, l'intervallo visualizzato sull'asse delle $x$ e delle $y$.

```r
hist(rnorm(1000), probability = TRUE, xlim=c(-4,4), ylim=c(0,.5)) # istogramma delle densità
curve(dnorm(x), add = TRUE)
```

### Box plot

Per definire un box-plot è necessario definire i quartili di un campione (vettore di dati) con una relazione d'ordine, ossia quei tre valori che ripartiscono in quattro parti di uguale numerosità il campione. Li trovi denotati tipicamente con $q_1$, $q_2$ e $q_3$. $q_2$ è anche detto *mediana* o *valore mediano*.

Il box plot è dunque un rettangolo  (orientato orizzontalmente o verticalmente) diviso in due parti, da cui escono due segmenti. Il rettangolo (la "scatola") è delimitato dal primo e dal terzo quartile, $q_1$ e $q_3$, e diviso al suo interno dalla mediana, $q_2$. I segmenti (i "baffi") sono delimitati dal minimo e dal massimo del campione.

```r
boxplot(df$mpg)
```

Analogamente, puoi visualizzare il box plot relativo ad `n`=1000 osservazioni di una v.a. normale standard.

```r
boxplot(rnorm(1000))
```

### Grafici: più di due variabili (numeriche)

Per esplorare un data set con più di due variabili numeriche è utile esplorare una matrice di grafici a dispersione (*scatter plot matrix*) tra tutte le coppie di variabili.
Puoi utilizzare il comando di base `plot()`, che se applicato ad un argomento di classe `"data.frame"`, con più di due colonne, restituisce automaticamente la scatter plot matrix tra tutte le variabili numeriche presenti nel data set.

```r
pairs(df[, 3:5])
```

Puoi inoltre colorare i punti in base ad una variabile, con un numero finito di valori/livelli, come per esempio `df$origin` nell'esempio seguente che è di classe `integer` ed assume i valori {1,2,3}. Per farlo usa l'argomento `col=`. 

```r
pairs(df[, 3:5], col = df$origin)
```

Tieni presente che R codifica ogni colore con un numero. Per esempio, al numero 1 è associato il nero, al 2 il rosso, allo 0 il bianco . Anche TRUE e FALSE sono codificati con 1 e 0 rispettivamente, quindi `col=FALSE` o `col=F` equivale a `col=0` ossia `col="white"`, quindi fai attenzione!

```r
pairs(df[, 3:5], col = FALSE)
```
