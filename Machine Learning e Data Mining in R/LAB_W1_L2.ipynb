## Data transformation: `dplyr`

Principali funzioni:
* `filter()`, filtra le osservazioni secondo i propri valori
* `arrange()`, riordina le righe
* `select()`, seleziona le variabili/colonne mediante il proprio nome 
* `mutate()`, crea nuove variabili come funzioni di variabili già esistenti
* `summarise()`, sintetizza le informazioni contenute in una o più colonne
* `group_by()`, stratifica le analisi

Tutte queste funzioni hanno la stessa sintassi: 
* Il primo argomento è il data frame da trasformare
* I successivi argomenti sono i nomi delle variabili (senza virgolette) su cui operare
* Il risultato è un nuovo data frame

### L'operatore pipe `%>%`

* Questo operatore è molto utile per scrivere codici in maniera ordinata e facili da leggere, quando vuoi applicare differenti funzioni in maniera sequenziale. 
* Scorciatoia da tastiera: CTRL/CMD+SHIFT+M.

Per capirne l'uso ricorriamo subito ad un esempio:

```r
x <- 1:10
sum(x)
x %>% sum(.) ## è equivalente a sum(x)
x %>% sum() ## se non usi un punto, x è usato come primo ARGOMENTO (nota: qui abbiamo solamente un argomento)
x %>% sum ## puoi anche evitare di usare le parentesi se non vuoi avere altri argomenti
```

L'operatore `%>%` si può leggere *quindi* (traduzione dall'inglese di *then*).  `x%>%sum` si potrebbe leggere *prendi gli elementi del vettore x, quindi sommali*.  Considera un secondo esempio, con l'applicazione di due funzioni consecutive. Supponi di voler calcolare la somma dei valori assoluti. Due modi equivalenti di farlo sono i seguenti.

```r
y <- c(- 5, 3, 8, - 1, 0.4)
sum(abs(y))
y %>% 
  abs %>% 
  sum ## è conveniente usare nuove righe di codice per ciascun pipe quando il codice diventa lungo
```

`y%>%abs%>%sum` si potrebbe leggere *prendi gli elementi del vettore x, quindi calcolane il valore assoluto, quindi sommali*. Probabilmente è ancora conveniente usare, per il momento, l'istruzione compatta codice `sum(abs(y))`, ma apprezzerai la leggibilità del codice scritto usando l'operatore pipe operando sui data frame, ad esempio per  filtrare righe,  selezionare colonne e sintetizzarne i risultati .

Quando invece vuoi passare più di un argomento ad una funzione, l'uso di pipe è il seguente: 

```r
z <- c(1, 5, NA, - 3.6)
mean(z, na.rm = TRUE)
z %>% mean(na.rm = TRUE)
```

Di seguito si riportano degli esempi per illustrare come l'argomento `.` interagisce con l'operatore pipe. 

```r
a <- 1
b <- 2
c(a, b)
a %>% c(., b) # equivalente a c(a, b)
a %>% c(b, .) # equivalente a c(b, a)
a %>% c(b)    # equivalente a c(a, b) NESSUN PUNTO: PRIMO ARGOMENTO
```

* Perché usare l'operatore pipe `%>%`? Come potrai vedere in seguito, la maggior parte delle funzioni di `dplyr`, come per le altre funzioni di `tidyverse`, sono adatte ad essere usate con `%>%`. 

* L'uso di  `%>%` è sempre da preferire? No, l'uso di  `%>%` è particolarmente vantaggioso con le funzioni del pacchetto `dplyr`, ma ci sono casi in cui non è di alcuna convenienza come, per esempio, quando usiamo delle espressioni matematiche.  

Supponiamo di voler calcolare la somma delle differenze dei quadrati di due vettori:

```r
a <- c(1, 4, 2)
b <- c(3, 7.2, - 9)
sum((a - b) ^ 2)
```

Potresti usare l'operatore `%>%` nel modo seguente, ed accorgerti che non dà alcun vantaggio. 

```r
a %>% 
  `-`(b) %>% 
  `^`(2) %>% 
  sum
```

### Estrarre righe sulla base di una condizione mediante la funzione `filter()`

Estrai i voli corrispondenti al 25 Marzo. Normalmente scriveresti

```r
flights[flights$month == 3 | flights$day == 25, ]
```

con `dplyr`:

```r
filter(flights, month == 3, day == 25)
```

Come molte altre funzioni in `dplyr`, `filter()` prevede in input un data frame come primo argomento che può essere fornito mediante l'uso dell'operatore `%>%`, per ottenere lo stesso risultato ma un codice più leggibile, quando bisogna concatenare più funzioni di questo tipo.

```r
flights %>% filter(month == 3, day == 25)
```

In caso di calcoli più complessi che necessitino della composizione di più funzioni, puoi assegnare il risultato ad una variabile e usare questa variabile come argomento di altre funzioni. Per esempio:

```r
Marzo25 <- filter(flights, month == 3, day == 25)
Marzo25
head(Marzo25)
tail(Marzo25)
```

Per evitare l'assegnazione strumentale di un risultato intermedio ad una variabile temporanea su cui applicare la funzione successiva puoi sempre comporre le funzioni nel modo seguente:

```r
head(filter(flights, month == 3, day == 25))
```

Ma il codice comincia ad essere di difficile lettura. Uno dei grandi vantaggi derivanti dall'uso dell'operatore `%>%`, è come sappiamo migliorare la leggibilità soprattutto nei pacchetti contenuti in `tidyverse`: 

```r
flights %>% 
  filter(month == 3, day == 25) %>% 
  head
```

### Riordinare le righe con `arrange()`

Se vuoi ordinare le righe di un data frame in R, potresti usare la funzione `order()`. Per esempio, supponi di voler riordinare le righe di `flights` secondo i valori di `arr_delay`:

```r
select(flights, starts_with("sched_"))
select(flights, ends_with("_time"))
select(flights, contains("arr_"))
## posiziona all'inizio le colonne origin e dest, poi lascia invariata la posizione delle altre colonne
```

### Sintetizzare le informazioni contenute in una o più colonne con `summarise()`

Se volessi sintetizzare le informazioni contenute in una o più colonne di un data frame in una singola riga scriveresti un'istruzione del tipo:

### Stratificare le analisi con  `group_by()`

La funzione `group_by()` permette la stratificazione del data set, ossia di raggruppare le righe del data frame in base ai valori assunti da una variabile. 

### Uso congiunto di  `group_by()` + `summarise()`

L'utilizzo congiunto di `group_by()` con `summarise()` permette di applicare l'analisi di un dataset completo a quello di gruppi individuali. Se usi il comando `summarise()` dopo `group_by()`, come nell'esempio seguente:

```r
flights %>% 
  group_by(month) %>% 
  summarise(distanza_media = mean(distance, na.rm = TRUE))
```

equivale a dividere le righe del dataset in base ai gruppi definiti dalla colonna/variabile scelta (nel nostro caso il mese), per poi comprimere queste righe in una sola, in base alla funzione scelta (nel nostro caso la funzione `mean()`) e registrare i valori di sintesi nella colonna `distanza_media`. 
Puoi raggruppare le righe anche rispetto a più di una variabile, come nell'esempio seguente:

```r
flights %>% 
  group_by(year, month) %>% 
  summarise(monthly_dep_delay = mean(dep_delay, na.rm = TRUE))
```

Nota che `group_by()` stratifica in base a variabili *categoriche* o di tipo *factor* o vettori di caratteri che presentano tipicamente un numero limitato di valori detti anche *livelli*. 

### Uso congiunto di `group_by()`+ `mutate()` 

Se al posto di `summarise()` dopo `group_by()`, usi il comando `mutate()`, le informazioni sintetiche raggruppate sono associate alle corrispondenti righe di ciascun gruppo come una nuova colonna, e il numero di righe del dataset non viene modificato. 

Nell'esempio seguente, ti mostro l'uso di `mutate()` al posto di `summarise()` dopo `group_by()`

```r
flights %>% 
  group_by(origin) %>% 
  mutate(ritardo_medio_arrivo= mean(arr_delay, na.rm = TRUE))
```

in cui le singole statistiche raggruppate sono ripetute per tutte le righe in ciascun gruppo. 

Ti può essere conveniente combinare il risultato con `select()` per estrarre solo le eventuali colonne a cui sei interessato.

```r
flights %>% 
  group_by(origin) %>% 
  mutate(ritardo_medio_arrivo = mean(arr_delay, na.rm = TRUE)) %>% 
  select(origin, arr_delay, ritardo_medio_arrivo)
```

### Uso di  `n()` e `count()`

Una funzione di sintesi che si può usare all'interno di `summarise()`  (per esempio al posto di `mean()`) è `n()`, che restituisce la dimensione di ciascun gruppo considerato, come nell'esempio seguente

```r
flights %>% 
  group_by(month, day) %>% 
  summarise(n = n())
```

Tuttavia, c'è un modo più semplice per ottenere lo stesso risultato attraverso la funzione `count()` di  `dplyr`  

```r
flights %>% 
  group_by(month, day) %>% 
  count
```

### Uso di `group_by()` con variabili di tipo numerico

Per poter raggruppare una variabile numerica con un elevato numero di valori differenti mediante il comando  `group_by()` hai bisogno di definire degli intervalli e stratificare in base a questi (come avviene per le classi di un istogramma).

Puoi creare una nuova colonna con il comando `mutate()`, definendo una variabile booleana (VERO/FALSO) o usando `ifelse()`, se devi definire solo due intervalli, altrimenti usando  `cut()` o `case_when()`. 
Nell'esempio che segue, ti mostro come contare il numero di righe nel primo e nel secondo semestre dell'anno.  Visto che ci sono solo due condizioni, puoi creare una variabile booleana

```r
flights %>% 
  mutate(primo_semestre = month <= 6) %>% 
  group_by(primo_semestre) %>% 
  count
```

o alternativamente usare `ifelse()` per creare una variabile categorica a due livelli. 

```r
flights %>% 
  mutate(semestre = ifelse(month <= 6, "primo", "secondo")) %>% 
  group_by(semestre) %>% 
  count
```

Con più di due livelli, potresti annidare più comandi `ifelse()`  

```r
flights %>%
  mutate(trimester = ifelse(
    month <= 3, "primo", ifelse(
      month <= 6, "secondo", ifelse(
        month <= 9, "terzo", "quarto")))) %>%
  group_by(trimester) %>%
  count
```

ma ti risulterà difficile scrivere un codice compatto e leggibile. Attraverso `case_when()` di `dplyr` puoi avere lo stesso risultato.

```r
flights %>% 
  mutate(trimester = case_when(
    month <= 3 ~ "primo",
    month >  3 & month <= 6 ~ "secondo",
    month >  6 & month <= 9 ~ "terzo",
    month >  9 ~ "quarto")) %>% 
  group_by(trimester) %>% 
  count
```

Il codice non è breve, ma è più facile da leggere e comprendere.

### `tidyr`: pivoting

La funzione più comunemente usata nel pacchetto `tidyr` contenuto in `tidyverse` è `pivot_longer()`, che è la nuova versione del comando `gather()`.  `pivot_longer()` trasforma un dataframe largo e corto in uno stretto e lungo, detto in formato `tidy`. Ti mostro, senza giri di parole, come opera il comando `pivot_longer()` attraverso l'esempio seguente.

```r
table4a
table4a %>% 
  pivot_longer(- country, names_to = "year", values_to = "cases")
```

L'utilità di questo comando sarà chiara nella prossima lezione in cui introdurremo la visualizzazione dei dati attraverso il pacchetto `ggplot2`.

```r
table4a %>% 
  pivot_longer(- country, names_to = "year", values_to = "cases") %>% 
  mutate(year = as.numeric(year)) %>% 
  ggplot +
  geom_line(aes(year, cases, col = country))
```
